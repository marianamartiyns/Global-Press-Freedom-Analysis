```python
import pandas as pd 
import numpy as np
import seaborn as sns

df = pd.read_csv('World Press Freedom Score.csv')

df = df.set_index('Country')

df
```

### Gráfico 1:
#### Serve como forma de análise para responder a pergunta 1
- Percebe-se que quanto maior o nivel de segurança maior é o indice de liberdade, com alguns dados destoando do restante.

```python
sns.scatterplot(data=df, x='Global Score 2022', y='Safety Score 2022', hue='Ranking', palette="rocket");
```

### Gráfico 2:
#### Relação de informações disponiveis do Brasil
- É possível perceber um grande aumento no global score de 2022, no contexto político-social de 2022 podemos ter uma ideia do por que desse crescimento ter ocorrido. 

```python
a = pd.DataFrame(df.loc['Brazil']).T

#dropei a coluna de abuse score pois nela não constava registro nenhum
a = a.drop(columns = ['Ranking','Abuse Score 2021'])
ax = sns.barplot(data = a, orient = 'h', palette = 'flare' )
```

### Gráfico 3:
#### Histograma dos dados de Contexto político de 2022
- Percebe-se uma concentração maior em valores aproximados de 60%
